- 프로젝트 구텐베르크(Project Gutenberg)에서 '로미오와 줄리엣' 텍스트 다운로드하여 사용

```python
import networkx as nx
import pandas as pd
import matplotlib.pyplot as plt
import nltk
from nltk.tokenize import sent_tokenize
from collections import defaultdict
import requests

# NLTK 데이터 다운로드
nltk.download('punkt')
```

```python
# 1. 텍스트 데이터 로드
url = 'https://gutenberg.org/files/1112/1112-0.txt'
response = requests.get(url)
text = response.text

text
```

```python
# 2. 등장인물 리스트 작성
characters = ['ROMEO', 'JULIET', 'MERCUTIO', 'TYBALT', 'BENVOLIO', 'CAPULET', 'LADY CAPULET', 'MONTAGUE', 'LADY MONTAGUE', 'FRIAR LAURENCE', 'NURSE', 'PARIS', 'PRINCE']
```

```python
# 3. 등장인물 간 상호작용 추출
# 대사 부분만 추출하여 처리
lines = text.split('\n')
dialogues = []
for line in lines:
    # 등장인물 분리 기준 다시 설정
    if line.startswith('   '):
        current_character = line.split('.')[0]
    elif line and not line.isupper():
        dialogues.append((current_character, line))
```
